```
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.animation as animation
import matplotlib.pyplot as plt
import random
```

```
samp = []
samp_avg = []
for roll in range(1,1000):
    x = np.random.randint(low=0,high=2, size=1)[0]
    samp.append(x)
    samp_avg.append(np.mean(samp))
```

```
def plot_sample (samp_avg,samp_no, subplot_no):
    title_font = {'family': 'arial', 'color': 'darkred','weight': 'bold','size': 13 }
    axis_font  = {'family': 'arial', 'color': 'darkblue','weight': 'bold','size': 10 }
    plt.subplot(subplot_no)
    plt.hlines(0.5, 0, samp_no, colors= 'r' )
    plt.plot(samp_avg[0:samp_no])
    plt.ylim((0,2))
    plt.title("Number of samples = {}".format(samp_no), fontdict = title_font)
    plt.xlabel("Number of samples", fontdict = axis_font)
    plt.ylabel("Sample Average", fontdict = axis_font)
    return plt
```

```
plt.figure(figsize=(15,3), dpi=100)
plot_sample(samp_avg,100, 131) 
plot_sample(samp_avg, 500, 132)
plot_sample(samp_avg, 2000, 133)
plt.show()
```

The main factor to afect the outcome is the number of samples taken. As seen from the outcomes of different sample sizes the more samples there are closer it is to the expected value.
